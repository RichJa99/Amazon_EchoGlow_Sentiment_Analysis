## Load Data

```python
df = pd.read_csv('amazon_EchoGlow_sentiment.csv')
df
```

```python
helpfull_review = df[df['Review Usefulness'] != 'No Data']
helpfull_review['Review Usefulness'] = helpfull_review['Review Usefulness'].str.split(' ')

Usefulness = []
for i in helpfull_review['Review Usefulness']:
    i = i[0]
    Usefulness.append(i)

helpfull_review['Review Usefulness'] = Usefulness
helpfull_review
```
